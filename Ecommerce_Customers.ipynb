```python
y=df['Yearly Amount Spent']
y
```

```python
x=df[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# **Model Building & Training**

**Model Training**

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
# Feature names
features = ["Avg. Session Length", "Time on App", "Time on Website", "Length of Membership"]

# Coefficients from the model
coefficients = [25.5962591, 38.78534598, 0.31038593, 61.89682859]

# Create a DataFrame to pair features with coefficients
coef_df = pd.DataFrame({
    'Feature': features,
    'Coefficient': coefficients
})

print(coef_df)
```

```python
# Add interpretation
coef_df['Interpretation'] = [
    "For every additional unit of Avg. Session Length, the target variable increases by 25.60 units.",
    "For every additional unit of Time on App, the target variable increases by 38.79 units.",
    "For every additional unit of Time on Website, the target variable increases by 0.31 units.",
    "For every additional unit of Length of Membership, the target variable increases by 61.90 units."
]

print(coef_df)
```

```python
# Plotting the coefficients
plt.figure(figsize=(10, 6))
plt.barh(coef_df['Feature'], coef_df['Coefficient'], color='skyblue')
plt.xlabel('Coefficient Value')
plt.title('Feature Coefficients')
plt.show()
```
